# 03 Data Wrangling

```python
# change the currency to the same unit
# split the price column
df['Currency'] = df.loc[:,'Price'].str.slice(0,3)
```

```python
# count of curreny used
df['Currency'].value_counts(dropna = False)
```

```python
# split the price 
df['Price'] = df['Price'].str.replace(r'[A-Z\D\W]', '')
```

```python
df['Price'] = pd.to_numeric(df['Price'])
```

```python
# change data type of price to interger
df.dtypes
```

```python
# convert currency
def eurprice(Price,Currency):
    if Currency=='CHF':
        return Price * 0.99
    elif Currency=='DKK':
        return Price * 0.13
    elif Currency=='Â£':
        return Price * 1.14
    else:
        return Price
```

```python
df['EUR_price'] = df.apply(lambda x: eurprice(x.Price, x.Currency), axis=1)
```

```python
#check the dataframe
df.head()
```

```python
# extract Country and City from Location
df[['Country','City','Standing_point']]= df['Location'].str.split('Â»', 2, expand=True)
```

```python
# check the dataframe
df.head()
```

```python
# Drop the redundant columns
df = df.drop(['Price','Currency','Location','Standing_point'],axis = 1)
```
